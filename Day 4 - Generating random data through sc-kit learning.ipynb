# Generating random data through sci-kit learning

When we are creating models it becomes necessary to have good data for building models. In case if we dint find data to train, we can make our own through sc-kit learning library.

```python
#Import pandas and plotting libraries to visualize data
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

## 1. Generate isotropic Gaussian blobs for clustering

```python
from sklearn.datasets import make_blobs

blob_data, blob_labels = make_blobs(n_samples=500,
                                    n_features=2,
                                    centers=6,
                                    cluster_std=.5)

plt.scatter(blob_data[:,0],
            blob_data[:,1], 
            c = blob_labels,
            cmap='viridis');
```

```python
blob_df = pd.DataFrame({'x-axis':blob_data[:,0],'y-axix':blob_data[:,1],'Label':blob_labels})
blob_df.head(5)
```

## 2. Create circles

```python
from sklearn.datasets import make_circles

circles_data, circles_labels = make_circles(n_samples=500,
                                      noise=.01,
                                      factor=.7)

plt.scatter(circles_data[:,0],
            circles_data[:,1], 
            c = circles_labels,
            cmap='viridis');
```

```python
circles_df = pd.DataFrame({'x-axis':circles_data[:,0],'y-axis':circles_data[:,1],'Class':circles_labels})
circles_df.head(5)
```

## 3. Regression data

```python
from sklearn.datasets import make_regression

regression_data, regression_values = make_regression(n_samples=500,
                                                     n_features=1,
                                                     n_informative=1,
                                                     noise=5)

plt.scatter(regression_data[:,0],
            regression_values,
            cmap='viridis');
```

```python
regression_df = pd.DataFrame({'Feature 1':regression_data[:,0],'Value':regression_values})
regression_df.head(5)
```

## 4. Classification data

```python
from sklearn.datasets import make_classification

classification_data, classification_class = make_classification(n_samples=50,
                                                                 n_features=5,
                                                                 n_informative=3,
                                                                 n_redundant=1,
                                                                 n_classes=2)

classification_df = pd.DataFrame({'Variable 1':classification_data[:,0],
                                  'Varialbe 2':classification_data[:,1],
                                  'Variable 3':classification_data[:,2],
                                  'Variable 4':classification_data[:,3],
                                  'Variable 5':classification_data[:,4],
                                  'Class':classification_class})
classification_df.head(5)
```

## 5.Moon dataset

```python
from sklearn.datasets import make_moons

moons_data, moons_labels = make_moons(n_samples=100,noise=0)

plt.scatter(moons_data[:,0],
            moons_data[:,1],
            c=moons_labels,
            cmap='viridis');
```

## 6.Multilabel classification

```python
from sklearn.datasets import make_multilabel_classification

multilabel_classification_data, multilabel_classification_classes = make_multilabel_classification(n_samples=100,
                                                                                                  n_features=5,
                                                                                                  n_classes=2,
                                                                                                  n_labels=3)

multilabel_classification_df = pd.DataFrame({'Feature 1':multilabel_classification_data[:,0],
                                             'Feature 2':multilabel_classification_data[:,1],
                                             'Feature 3':multilabel_classification_data[:,2],
                                             'Feature 4':multilabel_classification_data[:,3],
                                             'Feature 5':multilabel_classification_data[:,4],
                                             'Class 1':multilabel_classification_classes[:,0],
                                             'Class 2':multilabel_classification_classes[:,1],
                                             'Class 3':multilabel_classification_classes[:,1]})
multilabel_classification_df.head(5)
```
